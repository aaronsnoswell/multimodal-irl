
# Visualising the Stick-Breaking representation


```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm, beta
```

```python

K = 10
alpha = 1.0
G0 = norm(0, 1)
V = beta(1.0, alpha)

etas = G0.rvs(K)
Vs = V.rvs(K)
Vs[-1] = 1.0

sns.set()
plt.figure(figsize=(9, 6), dpi=300)

for i, (eta_i, V_i) in enumerate(zip(etas, Vs)):
    mag = V_i * np.prod(1.0 - Vs[0:i])
    g = plt.scatter(eta_i, mag, marker='o', color='tab:blue', label='G')

x = np.arange(-10, 10, 0.1)
g0 = plt.plot(x, G0.pdf(x), label='G0')

plt.legend(
    [g0[0], g],
    [
        r"$G_0 = \mathcal{N}(0, 1)$",
        r"$G$"
    ]
)
plt.title(rf"Log plot of G vs. G0, $\alpha = {alpha}, K = {K}$")
#plt.semilogy()
plt.show()
plt.close()
```
